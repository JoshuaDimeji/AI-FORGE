Let's create a contact manager with Python. I'll call this my_contacts

```python
# This is a dictionary of dictionaries, where the name is key and the value is another dictionary

my_contacts = {
    "Tony Elumelu": {"phone_number":"+2348055622222", "email":"elumelu@example.com", "tags":["friend","work"]},
    "Aliko Dangote": {"phone_number":"+2348100224477", "email":"dangote@example.com", "tags":["family"]},
    "Abdulsalam Rabiu": {"phone_number":"+2349077888555", "email":"rabiu@example.com", "tags":["friend"]},
    "Elon Musk": {"phone_number":"+177722288883", "email":"elon@example.com", "tags":["work"]},
    "Femi Otedola": {"phone_number":"+2347011122200", "email":"otedola@example.com", "tags":["friend"]},
    "Anthony Joshua": {"phone_number":"+2348033115577", "email":"antony@example.com", "tags":["family"]}
}
```

Showing the output and confirming the data type

```python
# To view all details in the dictionary
print(my_contacts)

# Let's confirm the data type 
print(type(my_contacts))
```

Displaying all contact name only. Here, I showed two ways by which this can be achieved

```python
# To display all contact name only, a for loop can be used here
for names in my_contacts:
    print(names)
```

```python
# Another alternative is to use the .keys() method in the dictionary
for names in my_contacts.keys():
    print(names)
```

Displaying full details for a specific contact by name. Here, a user input was requested.

```python
# Display full details for a specific contact by name 
name_input = input("Enter contact name to view details")  # Requests for user's input
name_input = name_input.strip()      # removes all trailing and leading white spaces
if name_input in my_contacts.keys():  # Check if name is in dictionary before
    print(f"Here are the details for {name_input}:") # prints out the details
    print(f"Phone number: {my_contacts[name_input]["phone_number"]}")  # retrieves and prints the phone number from the dictionary
    print(f"Email address: {my_contacts[name_input]["email"]}")    # retrieves and prints the email from the dictionary
    print(f"Tags are: {my_contacts[name_input]["tags"]}")        # # retrieves and prints the tags from the dictionary
else:
    print(f"Details for {name_input} are not found!")   # If input isn't found there, returns this message
```

Adding a new contact. A user input was requested and checked if it was already in the dictionary, and then added if it wasn't

```python
# Adding a new contact
new_name = input("Enter new contact to add")   # Requests name input from user

new_name = new_name.strip()  # trim off leading and rear white spaces
if new_name in my_contacts.keys():  # Checks if name is already in the dictionary
    print("Name already exists here!")  # Returns this if found there
else:
    new_phone = input("Enter phone number here:")   # get phone number
    new_email = input("Enter email here:")          # get email
    new_tags = input("Enter tags, separated by comma if more than one:").split(",")  # get tags from user and split if more than one
    # remove extra spaces in the tag using .strip() method
    new_tags = [tag.strip() for tag in new_tags]

    # now, we put all of the user input in a dictionary data type by connecting key-value
    my_contacts[new_name] = {
        "phone_number" : new_phone,
        "email" : new_email,
        "tags" : new_tags
    }
    print("Contact added successfully!")

print(f"Now, the updated contact list is:\n {my_contacts}")
    
    
```

Find all contacts with specific tag and show their names only

```python
# Finding all contacts with spcific tag(s)
tag_search = input("Which tag do you want to search for? There is friend, work and family:")
tag_search = tag_search.lower()   #convert input string to lowercase
tag_search = tag_search.strip()   #clean all leading and rear white spaces

print(f"Contacts with tag '{tag_search}':")
found = False      #it is not yet found, assign the bool False to it

for names, details in my_contacts.items():  # loops through both keys and values of my_contacts dictionary
    if tag_search in details["tags"]: # if the tag been searched for is found, it should return the names i.e keys
        print(names)
        found = True     # then found is now True

if not found:   # if it's not found, it should return this message
    print(f"No contact found with the tag: {tag_search}")
    
```

Get a set of all unique tags across all contacts

```python
# Getting a set of all unique tags used across all contacts
unique_tags = set()  # remember that a set has unique values without repetitions

for details in my_contacts.values():      # for loop to look through the values of my_contacts dictionary
    unique_tags.update(details["tags"])

print("\nAll Unique Tags:")
print(unique_tags)
```
